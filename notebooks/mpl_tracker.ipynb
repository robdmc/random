```python
class Tracker:
    """
    Make jupyter streaming plots using matplotlib.
    """
    def __init__(self, rows=1, cols=1):
        self.line_dict = {}
        self.rows = rows
        self.cols = cols

    def init(self):
        from matplotlib import pyplot as plt
        self.figure, self.ax_list = plt.subplots(self.rows, self.cols, figsize=(9, 5))
        if self.rows * self.cols == 1:
            self.ax_list = [self.ax_list]
        else:
            self.ax_list = self.ax_list.flatten()
        for ax in self.ax_list:
            ax.grid(True)
        # plt.show(block=False)

    def track(self, x, y, name='metric', subplot=0, plotspec='k-'):
        if not (0 <= subplot < len(self.ax_list)):
            raise ValueError(f'Must have 0 <= subplot < {len(self.ax_list)}')

        ax = self.ax_list[subplot]
        line = self.line_dict.get(name)
        if line is None:
            line = self.line_dict[name] = ax.plot(x, y, plotspec, label=name)[0]
            ax.legend(loc='best')
        else:
            line.set_data(x, y)
            ax.relim()  # Recompute the data limits
            ax.autoscale_view()  # Automatically scale the axes to fit the new data
        
        self.figure.canvas.draw()

        

tracker = Tracker(rows=2, cols=2)
tracker.init()

```

```python
t = np.linspace(0, 2 * np.pi, 300)
for t0 in list(t)[0:100]:
    # tracker.start_iteration()
    ysp = np.sin(t0 - t)
    ysm = -ysp
    ycp = np.cos(t0 - t)
    ycm = - ycp
    tracker.track(t, ysp, subplot=0, name='0 ysp', plotspec='b', )
    tracker.track(t, ysm, subplot=0, name='0 ysm', plotspec='r', )
    tracker.track(t, ysp, subplot=1, name='1 ysp', plotspec='b', )
    tracker.track(t, ysm, subplot=1, name='1 ysm', plotspec='r', )
    tracker.track(t, ysp, subplot=2, name='2 ysp', plotspec='b', )
    tracker.track(t, ysm, subplot=2, name='2 ysm', plotspec='r', )
    tracker.track(t, ysp, subplot=3, name='3 ysp', plotspec='b', )
    tracker.track(t, ysm, subplot=3, name='3 ysm', plotspec='r', )
    time.sleep(.01)
```

```python

@dataclasses.dataclass
class JaxMLE:
    model: Any
    learning_rate: float = .1
    callback: Any = None
    callback_every: int = 1
    kwarg_to_track: Any = None
    w: Any = dataclasses.field(init=False) 
    cov:  Any = dataclasses.field(init=False) 
    sigma: Any = dataclasses.field(init=False) 
    N: Any = dataclasses.field(init=False)
    result: Any = dataclasses.field(init=False)
    best_state: Any = dataclasses.field(init=False)
    max_iter: int = 500 
    max_iter_after_best: int = 100

    @dataclasses.dataclass
    class IterationState:
        iterations: Any = None
        losses: Any = None
        best_iterations: Any = None
        best_losses: Any = None
        params: Any = None
        kwargs: Any = None
        y: Any = None
        yhat: Any = None
    
    example = textwrap.dedent("""
    Here is the examples you want to see
    """)

    def _loss(self, w, target,  **kwargs):
        y = self.model(w, **kwargs)
        if y.shape != target.shape:
            raise ValueError('The shape of the target variable must match model output shape')
        return jnp.sum((y - target) ** 2)
        

    def fit(self, w0, target, **kwargs):
        w = jnp.array(w0).astype(jnp.float32)
        target = jnp.array(target)

        optimizer = optax.adam(self.learning_rate)
        state = optimizer.init(w)
        steps = []
        losses = []
        self.best_losses = []
        self.best_iterations = []
        best_loss = 1e30
        try:
            for nn in range(self.max_iter):
                grads = jax.grad(self._loss)(w, y, **kwargs)
                updates, state = optimizer.update(grads, state)
                w = optax.apply_updates(w, updates)
                steps.append(nn)
                loss_val = self._loss(w, y, **kwargs)
                yf = model(w, **kwargs)
                losses.append(loss_val)
                if loss_val < best_loss:
                    best_loss = loss_val
                    self.best_losses.append(best_loss)
                    self.best_iterations.append(nn)
                    self.best_state = self.IterationState(
                        iterations=steps.copy(),
                        losses=losses.copy(),
                        best_iterations=self.best_iterations.copy(),
                        best_losses=self.best_losses.copy(),
                        params=w.copy(),
                        kwargs=kwargs,
                        y=target,
                        yhat=yf
                    )
                if self.callback is not None:
                    if nn % self.callback_every == 0:
                        iter_state = self.callback(
                            self.IterationState(
                                iterations=steps,
                                losses=losses,
                                best_iterations=self.best_iterations,
                                best_losses=self.best_losses,
                                params=w,
                                kwargs=kwargs,
                                y=target,
                                yhat=yf
                            )
                        )

                last_best_iter = 0
                found_a_best = bool(self.best_iterations)
                if found_a_best:
                    last_best_iter = self.best_iterations[-1]
                iterations_since_best = nn - last_best_iter
                if iterations_since_best > self.max_iter_after_best:
                    if self.callback is not None:
                        self.callback(self.best_state)
                    break
            
        except KeyboardInterrupt:
            print('Keyboard interrupt. Retaining current and best state.', file=sys.stderr)
            
        self.w = self.best_state.params
        self.N = len(target)

        information_matrix = jax.hessian(self._loss, argnums=0)(self.w, target, **kwargs)
        sigma = jnp.std(target - self.model(self.w, **kwargs))
        
        self.cov = sigma ** 2 * jnp.linalg.inv(information_matrix)
            
        self.sigma = jnp.sqrt(jnp.diag(self.cov).squeeze())
    

    def _compute_error(self, **kwargs):

        @jax.jit
        def model_with_only_params(w):
            return model(w, **kwargs)
        
        J2 = self.N * jax.jacfwd(model_with_only_params)(self.sigma) ** 2
        
        sigma2 = J2 @ (self.sigma.reshape([-1, 1]) ** 2)
        sigma = jnp.sqrt(sigma2).squeeze()
        return sigma

    def predict(self, **kwargs):
        return self.model(self.w, **kwargs)

    def predict_and_error(self, **kwargs):
        y = self.predict(**kwargs)
        sigma = self._compute_error(**kwargs)
        return y, sigma


```

```python
def model(w, t):
    a = w[0]
    f = w[1]
    t0 = w[2]
    gamma = w[3]
    y = a * jnp.sin(2 * np.pi * f * (t - t0)) * jnp.exp(gamma * t)
    return y

def call_back(state):
    splotter.update(state.iterations, state.losses, subplot=0, name='loss', logy=True)
    if state.best_iterations:
        splotter.update(state.best_iterations, state.best_losses, subplot=0, name='bestloss', plotspec='r.', logy=True)
    splotter.update(state.kwargs['t'], state.y, subplot=1, name='data', plotspec='k.')
    splotter.update(state.kwargs['t'], state.yhat, subplot=1, name='fit', plotspec='r-')
    splotter.update(state.best_iterations[-100:], state.best_losses[-100:], subplot=2, name='recentloss', logy=True)
    
    

t = np.linspace(0, 2 * np.pi, 600)
a_true = 2.
f_true = .5
t0_true = 1 * np.pi
gamma_true = 1/3.
w = np.array([a_true, f_true, t0_true, gamma_true])
y_true = model(w, t)


y = y_true + 5.5 * np.random.randn(*y_true.shape)

w0 = jnp.array([10, .4, 3., .5])
fitter = JaxMLE(model, learning_rate=.09, max_iter=18270, callback=call_back, callback_every=100)
fitter.fit(w0, y, t=t)
yf, sigma = fitter.predict_and_error(t=t)

ax = ezr.figure()
ax.fill_between(t, yf + sigma, yf - sigma, color='grey', alpha=.5)
ax.plot(t, yf, 'k-')
ax.plot(t, y, 'r.', alpha=1)

```
